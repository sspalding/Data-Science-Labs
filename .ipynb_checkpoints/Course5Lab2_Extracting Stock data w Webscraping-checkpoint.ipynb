```python
# libraries
#!pip install pandas==1.3.3
#!pip install requests==2.26.0
!pip install bs4
!pip install html5lib==1.1 
!pip install lxml==4.6.4
#!pip install plotly==5.3.1
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# using webscraping to extract stock data
# use the request library to download the webpage and extract the text
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
data  = requests.get(url).text
```

```python
# parse the text into html using beautiful soup
soup = BeautifulSoup(data,'html5lib')
```

```python
# turn the html table into a pandas dataframe
netflix_data = pd.DataFrame(columns = ['Data','Open','High','Low','Close','Volume'])

# isolate the body of the table which contains the info
# loop through each row and fild all the column values for each row
for row in soup.find('tbody').find_all('tr'):
    col = row.find_all('td')
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    
    # append the data of each row to the table
    netflix_data = netflix_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)
```

```python
# print out the data frame
netflix_data.head()
```

```python
# use the pandas read html function using the url
read_html_pandas_data = pd.read_html(url)
```

```python
# or we can convert the beautiful soup object to a string
read_html_pandas_data = pd.read_html(str(soup))
```

```python
# take the first table in the list returned, bc there is only one table on the page
netflix_dataframe = read_html_pandas_data[0]
netflix_dataframe.head()
```

```python
# Use the requests library to download the webpage https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html. 
# Save the text of the response as a variable named html_data.
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
html_data = requests.get(url).text
```

```python
# Question 1 What is the content of the title attribute:
print(soup.title)
```

```python
# Using beautiful soup extract the table with historical share prices and store it into a dataframe named amazon_data. 
# The dataframe should have columns Date, Open, High, Low, Close, Adj Close, and Volume. 
# Fill in each variable with the correct data from the list col.

amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date =col[0].text
    Open =col[1].text
    high =col[2].text
    low =col[3].text
    close =col[4].text
    adj_close =col[5].text
    volume =col[6].text
    
    amazon_data = amazon_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)
```
