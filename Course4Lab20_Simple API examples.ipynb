### Fruitvice API Example

```python
# import libraries
import requests
import json
```

```python
# obtain API data
data = requests.get("https://www.fruityvice.com/api/fruit/all")
```

```python
# retrieve results using json.loads() function
results = json.loads(data.text)
```

```python
# convert json data into a pandas data frame
pd.DataFrame(results)
```

```python
# the nutrition column contains subcolumns so the results need to be flattened or normalized
df2 = pd.json_normalize(results)
df2
```

```python
# find the family and genus of a cherry
cherry = df2.loc[df2["name"] == 'Cherry']
(cherry.iloc[0]['family']) , (cherry.iloc[0]['genus'])
```

```python
# how many calories are contained in a banana
banana = df2.loc[df2['name'] == 'Banana']
banana.iloc[0]['nutritions.calories']
```
